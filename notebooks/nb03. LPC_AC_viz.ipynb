```python
import pandas as pd
ac_results_file = "AC_viz_nodes_results.csv"
df_ac = pd.read_csv(ac_results_file)

lpc_results_file = "LPC_viz_nodes_results.csv"
df_lpc = pd.read_csv(lpc_results_file)
```

```python
import json 
f = open('LPC_AC_post.json',) 
post_data = json.load(f) 
f.close() 
```

```python
import plotly.graph_objects as go

C_ac_depths  = df_ac.iloc[-1,:]*100
C_lpc_depths = df_lpc.iloc[-1,:]*100

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=post_data['depths'], y=C_ac_depths,
                    mode='lines',
                    name='AC'))
fig.add_trace(go.Scatter(x=post_data['depths'], y=C_lpc_depths,
                    mode='lines',
                    name='LPC'))
fig.add_shape(type='line',opacity=0.3,
                x0=0,y0=0.4,
                x1=500,y1=0.4,
                line=dict(color='Green',),
                xref='x', yref='y'
             )

fig.update_layout(title='Carbon vs Depth',
                  xaxis_title='Depth [micron]',
                  yaxis_title='Carbon concentration [%]',
                  width=800, height=500,
                  paper_bgcolor="LightSteelBlue",
                  yaxis=dict(titlefont=dict(size=15)),
                  xaxis=dict(titlefont=dict(size=15))
                 )

fig.add_annotation(x=100, y=0.4,
                   text="Initial carbon",
                   font=dict(size=10),
                  )
fig.show()
# fig.write_html("trialPlotly.html")
```

```python
import plotly.graph_objects as go


C_lpc_surface = df_lpc.iloc[:,0]*100
C_lpc_200 = df_lpc.iloc[:,200]*100
C_ac_200 =df_ac.iloc[:,200]*100

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=post_data['times'], y=C_lpc_surface,
                    mode='lines',
                    name='LPC surface'))
fig.add_trace(go.Scatter(x=post_data['times'], y=C_lpc_200,
                    mode='lines',
                    name='LPC 200micron'))
fig.add_trace(go.Scatter(x=post_data['times'], y=C_ac_200,
                    mode='lines',
                    name='AC 200micron'))

fig.add_shape(type='line',opacity=0.3,
                x0=0,y0=0.4,
                x1=1800,y1=0.4,
                line=dict(color='Green',),
                xref='x', yref='y'
             )

fig.update_layout(title='Carbon vs Time',
                  xaxis_title='Time [s]',
                  yaxis_title='Carbon concentration [%]',
                  width=800, height=500,
                  paper_bgcolor='rgba(0,0,0,0)',
                  yaxis=dict(titlefont=dict(size=15)),
                  xaxis=dict(titlefont=dict(size=15))
                 )

fig.add_annotation(x=1000, y=0.4,
                   text="Initial carbon",
                   font=dict(size=10),
                  )
fig.show()
# fig.write_html("trialPlotly.html")
```
